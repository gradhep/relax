```python
import jax
import jax.numpy as jnp
from jax.random import PRNGKey, multivariate_normal
import matplotlib.pyplot as plt

# generate a 2D gaussian blob
def generate_blob(
    key=PRNGKey(0),
    n_samples=1000,
    mu=jnp.array([1, 1]),
    sigma=jnp.array([[1, 0.5], [0.5, 1]]),
):
    key, subkey = jax.random.split(key)
    return multivariate_normal(key, mu, sigma, (n_samples,))


data = generate_blob()
bins_2d = jnp.meshgrid(jnp.linspace(-3, 3, 10), jnp.linspace(-3, 3, 10))
plt.hist2d(data[:, 0], data[:, 1], bins=bins_2d)
```
